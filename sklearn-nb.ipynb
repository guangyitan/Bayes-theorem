```python
# Make Predictions with Naive Bayes On The Iris Dataset
from csv import reader
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load a CSV file
def load_csv(filename):
	count = 0
	dataset = list()
	with open(filename, 'r', encoding="utf8") as file:
		csv_reader = reader(file)
		for row in csv_reader:
			# remove column names
			if count == 0:
				count = 1
				continue
			dataset.append(row)
	return dataset

# convert categorical value to numerical
def convert_columns_to_numerical(dataset):
    for row in dataset:
        # Female->0, Male->1
        if row[0] == "Female":
            row[0] = 0
        else:
            row[0] = 1
        
        # No->0, Yes-> 1
        if row[-1] == "NO":
            row[-1] = 0
        else:
            row[-1] = 1
    return dataset

def string_to_int(dataset):
	for row in dataset:
		for i in range(len(row)):
			if isinstance(row[i], int):
				continue
			else:
				try:
					row[i] = int(row[i])
				except ValueError:
					continue
	return dataset

# Load the CSV file into a Pandas DataFrame
# df = pd.read_csv('dataset.csv')

dataset = load_csv("dataset.csv")
dataset = convert_columns_to_numerical(dataset)
dataset = string_to_int(dataset)

df = pd.DataFrame(dataset)
# Split the dataset into features and target variable
X = df.iloc[:, 1:]
y = df.iloc[:, 0]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Train the classifier on the training data
# gnb.fit(X_train, y_train)
gnb.fit(X,y)

# Make predictions on the testing data
y_pred = gnb.predict(X)

# Evaluate the accuracy of the predictions
accuracy = accuracy_score(y, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))

from pandas import DataFrame
from matplotlib import pyplot
dataset = DataFrame(dataset)
# summarize
print(dataset.describe())
# histograms of the variables
dataset.hist()
pyplot.show()

from sklearn.preprocessing import StandardScaler

# create an instance of the StandardScaler class
scaler = StandardScaler()

# fit the scaler to the data and transform it
X_train_scaled = scaler.fit_transform(X)

# Create a Gaussian Naive Bayes classifier
gnb2 = GaussianNB()

# Train the classifier on the training data
# gnb.fit(X_train, y_train)
gnb2.fit(X_train_scaled,y)

# Make predictions on the testing data
y_pred = gnb2.predict(X_train_scaled)

# Evaluate the accuracy of the predictions
accuracy = accuracy_score(y, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))

from pandas import DataFrame
from matplotlib import pyplot
dataset = DataFrame(X)
# summarize
print(dataset.describe())
# histograms of the variables
dataset.hist()
pyplot.show()
```

# iris dataset

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the iris dataset
iris = load_iris()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Create a Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Train the classifier on the training data
gnb.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = gnb.predict(X_test)

# Evaluate the accuracy of the predictions
accuracy = (y_pred == y_test).sum() / len(y_test)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
